```
import pandas as pd
```

```
df = pd.read_csv('https://raw.githubusercontent.com/jenfly/opsd/master/opsd_germany_daily.csv',parse_dates=True)
```

```
df['Date'] = pd.to_datetime(df['Date'])
```

```
df.dtypes
```

```
df.set_index('Date')
```

```
df = df.set_index('Date')
```

```
df['Year'] = df.index.year
df['Month'] = df.index.month
df['day'] = df.index.day
```

```
df.loc['2016-12']
```

```
import matplotlib.pyplot as plt
import seaborn as sns
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(11, 4)})
```

```
cols_plot = ['Consumption', 'Solar', 'Wind']
axes = df[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('Daily Totals (GWh)')
```

```
df.loc['2015']['Consumption'].plot(marker = '.', linestyle = None )
```

```
import matplotlib.dates as mdates
fig, ax = plt.subplots()
ax.plot(df.loc['2017-01':'2017-02', 'Consumption'], marker='o', linestyle='-')
ax.set_ylabel('Daily Consumption (GWh)')
ax.set_title('Jan-Feb 2017 Electricity Consumption')
# Set x-axis major ticks to weekly interval, on Mondays
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MONDAY))
# Format x-tick labels as 3-letter month name and day number
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'));
```

```
fig, axes = plt.subplots(3, 1, figsize=(11, 10), sharex=True)
for name, ax in zip(['Consumption', 'Solar', 'Wind'], axes):
  sns.boxplot(data=df, x='day', y=name, ax=ax)
  ax.set_ylabel('GWh')
  ax.set_title(name)
# Remove the automatic x-axis label from all but the bottom subplot
  if ax != axes[-1]:
      ax.set_xlabel('')
```

```
fig, axes = plt.subplots(3, 1, figsize=(11, 10), sharex=True)
for name, ax in zip(['Consumption', 'Solar', 'Wind'], axes):
  df.groupby('Month')[name].mean().plot(ax =ax)
  ax.set_ylabel('Mean GWh')
  ax.set_title(name)
# Remove the automatic x-axis label from all but the bottom subplot
  if ax != axes[-1]:
      ax.set_xlabel('')
```

```
df.groupby('Month')['Consumption'].median().plot()
```

```
df['Consumption'] = df['Consumption'].fillna(method='ffill')
df['Wind']=df['Wind'].fillna(method ='bfill' )
df['Solar'] = df['Solar'].fillna(method='bfill')
df['Wind+Solar'] = df['Wind+Solar'].fillna(method = 'bfill')
```

```
df_freq = df.asfreq('D',method='ffill')
```
